```python
from keras.applications.vgg16 import VGG16
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input, decode_predictions
import numpy as np
```

```python
model = VGG16(weights='imagenet')
```

```python
img = image.load_img('C:/Users/taleb/Pictures/image1.jpg',target_size=(224,224))
x = image.img_to_array(img)
```

```python
def predict(frame_224):
    x = np.expand_dims(frame_224, axis =0)
    x = preprocess_input(x)
    p = model.predict(x)
    return decode_predictions(p , top =1)[0][0][1]
```

```python
img = image.load_img('C:/Users/taleb/Pictures/image1.jpg',target_size=(224,224))
x = image.img_to_array(img)

predict(x)
```

```python
import cv2
import matplotlib.pyplot as plt

cap = cv2.VideoCapture(0)
while True:
    _, frame = cap.read()
    frame_224 = cv2.resize(frame,(224,224))
                          
    result = predict(frame_224)
    cv2.putText(frame, result,(40,50),cv2.FONT_HERSHEY_SIMPLEX,
               2,(0,255,0),3,cv2.LINE_AA)
    cv2.imshow("live camera :" , frame)
    if cv2.waitKey(1)==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
